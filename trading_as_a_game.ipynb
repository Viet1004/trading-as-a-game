```python
## Save train data
AUDNZD = pd.read_csv("Data/AUDNZD.csv")
AUDUSD = pd.read_csv("Data/AUDUSD.csv")
CADJPY = pd.read_csv("Data/CADJPY.csv")
CHFJPY = pd.read_csv("Data/CHFJPY.csv")

AUDNZD.drop("Unnamed: 0", axis=1, inplace=True)
AUDUSD.drop("Unnamed: 0", axis=1, inplace=True)
CADJPY.drop("Unnamed: 0", axis=1, inplace=True)
CHFJPY.drop("Unnamed: 0", axis=1, inplace=True)

AUDNZD.interpolate(axis=0, inplace=True)
AUDUSD.interpolate(axis=0, inplace=True)
CADJPY.interpolate(axis=0, inplace=True)
CHFJPY.interpolate(axis=0, inplace=True)

data = [AUDNZD, AUDUSD, CADJPY, CHFJPY]
```

```python
## Save test data
AUDNZD_test = pd.read_csv("Data/AUDNZD_test.csv")
AUDUSD_test = pd.read_csv("Data/AUDUSD_test.csv")
CADJPY_test = pd.read_csv("Data/CADJPY_test.csv")
CHFJPY_test = pd.read_csv("Data/CHFJPY_test.csv")

AUDNZD_test.drop("Unnamed: 0", axis=1, inplace=True)
AUDUSD_test.drop("Unnamed: 0", axis=1, inplace=True)
CADJPY_test.drop("Unnamed: 0", axis=1, inplace=True)
CHFJPY_test.drop("Unnamed: 0", axis=1, inplace=True)

AUDNZD_test.interpolate(axis=0, inplace=True)
AUDUSD_test.interpolate(axis=0, inplace=True)
CADJPY_test.interpolate(axis=0, inplace=True)
CHFJPY_test.interpolate(axis=0, inplace=True)

data_test = [AUDNZD_test, AUDUSD_test, CADJPY_test, CHFJPY_test]
```

```python
exchange_list = ["AUDNZD", "AUDUSD", "CADJPY", "CHFJPY"]
count = 0
for exchange in data:
    exchange.plot(x="time", y=["open", "close"], figsize=(20,15))
    plt.savefig(f"{exchange_list[count]}.png")
    count += 1
```
